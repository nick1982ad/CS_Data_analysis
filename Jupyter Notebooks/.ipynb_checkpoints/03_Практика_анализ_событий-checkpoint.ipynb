### 1.2. Нормализация с помощью pygrok

Сырое поле `raw_message` имеет формат, в котором Windows EventLog отображает событие в текстовом виде. Эти строки удобно нормализовать шаблонами grok.

Здесь мы задаём паттерны для интересующих нас EventID и вытаскиваем структурированные поля.

Список Event ID, для которых проводится анализ:

| Event ID| Описание    |
|---------|-------------|
| 4624    | Успешный вход пользователя |
| 4625    | Неуспешная попытка входа   |
| 4634    | Выход пользователя из системы |
| 4672    | Присвоение особых привилегий новому входу |
| 4688    | Создан новый процесс |
| 4663    | Попытка доступа к объекту |

````python
from pygrok import Grok

CUSTOM_PATTERNS = {
    "WINUSER":   r"[A-Za-z0-9_.\-$]+",
    "WINDOMAIN": r"[A-Za-z0-9_.\-]+",
    "HEXID":     r"0x[0-9A-Fa-f]+",
    "GUIDX":     r"\{[0-9A-Fa-f\-]{36}\}",
    "IPADDR":    r"(?:\d{1,3}\.){3}\d{1,3}|::1|-",
}

GROK_4624 = Grok(
    r"An account was successfully logged on\.\s+"
    r"Subject:\s+Account Name:\s+%{WINUSER:subject_user}\s+Logon ID:\s+%{HEXID:subject_logon_id}\s+"
    r"New Logon:\s+Account Name:\s+%{WINUSER:target_user}\s+Account Domain:\s+%{WINDOMAIN:target_domain}\s+"
    r"Logon ID:\s+%{HEXID:target_logon_id}\s+Logon GUID:\s+(?:%{GUIDX:logon_guid}|-)\s+"
    r"Logon Type:\s+%{INT:logon_type:int}\s+"
    r"Network Information:\s+Workstation Name:\s+%{WINUSER:workstation}\s+Source Network Address:\s+%{IPADDR:src_ip}",
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_4625 = Grok(
    r"An account failed to log on\.\s+Account Name:\s+%{WINUSER:target_user}\s+Account Domain:\s+%{WINDOMAIN:target_domain}\s+"
    r"Failure Information:\s+Sub Status:\s+%{HEXID:sub_status}\s+Logon Type:\s+%{INT:logon_type:int}\s+"
    r"Source Network Address:\s+%{IPADDR:src_ip}",
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_4634 = Grok(
    r"An account was logged off\.\s+Account Name:\s+%{WINUSER:target_user}\s+Logon ID:\s+%{HEXID:target_logon_id}\s+Logon Type:\s+(?:%{INT:logon_type:int}|-)",
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_4672 = Grok(
    r"Special privileges assigned to new logon\.\s+Account Name:\s+%{WINUSER:subject_user}\s+Logon ID:\s+%{HEXID:subject_logon_id}\s+Privileges:\s+%{GREEDYDATA:privileges}",
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_4688 = Grok(
    r"A new process has been created\.\s+Creator Subject:\s+Account Name:\s+%{WINUSER:subject_user}\s+Logon ID:\s+%{HEXID:subject_logon_id}\s+"
    r"Process Information:\s+New Process Name:\s+%{GREEDYDATA:process_name}\s+Command Line:\s+%{GREEDYDATA:cmdline}\s+Creator Process Name:\s+%{GREEDYDATA:parent_process}",
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_4663 = Grok(
    r"An attempt was made to access an object\.\s+Account Name:\s+%{WINUSER:subject_user}\s+Logon ID:\s+%{HEXID:subject_logon_id}\s+"
    r"Object Name:\s+%{GREEDYDATA:object_name}\s+Access Mask:\s+%{HEXID:access_mask}",
    custom_patterns=CUSTOM_PATTERNS,
)

GROK_DISPATCH = {
    4624: GROK_4624, 4625: GROK_4625, 4634: GROK_4634,
    4672: GROK_4672, 4688: GROK_4688, 4663: GROK_4663,
}
````

````python
def normalize_event(ev):
    """Возвращает плоский dict с унифицированной схемой полей.

    """
    eid = int(ev.get("EventID",0))
    out = {
        "ts": pd.to_datetime(ev.get("@timestamp")),
        "host": ev.get("Hostname"),
        "channel": ev.get("Channel", "Security"),
        "event_id": eid,
        "raw_message": ev.get("raw_message"),
        "target_user": None, "target_domain": None, "target_logon_id": None,
        "subject_user": None, "subject_logon_id": None,
        "logon_type": None, "src_ip": None, "workstation": None, "logon_guid": None,
        "process_name": None, "cmdline": None, "parent_process": None,
        "privileges": None, "object_name": None, "access_mask": None,
        "sub_status": None,
    }

    grok = GROK_DISPATCH.get(eid)
    if grok is not None and out["raw_message"]:
        m = grok.match(out["raw_message"])
        if m:
            for k, v in m.items():
                out[k] = v

    return out

normalized = [normalize_event(e) for e in raw_events]
df = pd.DataFrame(normalized).sort_values("ts").reset_index(drop=True)
print(f"Нормализовано событий: {len(df)}")
df.head(3).T
````

### 1.4. Расширение шаблонов grok на EventID 4720 и 4732

В реальной атаке после получения привилегий злоумышленник часто **создаёт «закладочную» учётную запись** и добавляет её в группу `Administrators`. Это даёт ему вход даже после смены пароля захваченного администратора. Такие действия демаскируются двумя EventID:

* **4720** — *«A user account was created»*. Поля: `New Account` (созданный пользователь), `Subject` (кто создал).
* **4732** — *«A member was added to a security-enabled local group»*. Поля: `Member` (добавленный SID/имя), `Group Name`.

Расширим словарь pygrok соответствующими типами событий.

````python
# 1.4.2. Регистрируем grok-паттерны для 4720 и 4732
GROK_4720 = Grok(
    r'A user account was created\.\s+'
    r'Subject:\s+Account Name:\s+%{WINUSER:subject_user}\s+Logon ID:\s+%{HEXID:subject_logon_id}\s+'
    r'New Account:\s+Account Name:\s+%{WINUSER:target_user}\s+Account Domain:\s+%{WINDOMAIN:target_domain}',
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_4732 = Grok(
    r'A member was added to a security-enabled local group\.\s+'
    r'Subject:\s+Account Name:\s+%{WINUSER:subject_user}\s+Logon ID:\s+%{HEXID:subject_logon_id}\s+'
    r'Member:\s+Account Name:\s+%{GREEDYDATA:member_name}\s+'
    r'Group:\s+Group Name:\s+%{WINUSER:group_name}',
    custom_patterns=CUSTOM_PATTERNS,
)
GROK_DISPATCH[4720] = GROK_4720
GROK_DISPATCH[4732] = GROK_4732

# Расширим нормализатор полем member_name/group_name
def normalize_event_v2(ev):
    out = normalize_event(ev)
    out.setdefault('member_name', None)
    out.setdefault('group_name', None)
    eid = out['event_id']
    grok = GROK_DISPATCH.get(eid)
    if grok is not None and out['raw_message']:
        m = grok.match(out['raw_message'])
        if m:
            for k, v in m.items():
                if k in ('member_name','group_name') and out.get(k) in (None,'','-'):
                    out[k] = v
    return out

extra_normalized = [normalize_event_v2(e) for e in raw_events]
df = pd.DataFrame(extra_normalized)

df = df.sort_values('ts').reset_index(drop=True)
````

## 2. Разбиение потока событий на отдельные сеансы с помощью идентификатора `LogonId`

В Windows Security есть два связанных параметра в каждом событии:

* **TargetLogonId** — ID создаваемого/завершаемого сеанса (события `4624`, `4634`, `4625`).
* **SubjectLogonId** — ID сеанса, в котором событие произошло (`4672`, `4688`, `4663` и т.п.).

Чтобы получить «сеанс пользователя» в рамках одного хоста, мы привязываем все события к одному **session key** = `(host, logon_id)`. Сеанс открывает `4624`, закрывает `4634`. Между ними попадают `4672` (выдача привилегий), `4688` (запуск процессов), `4663` (доступ к объектам).

````python
def assign_session_key(row):
    host = row["host"]
    if row["event_id"] in (4624, 4625, 4634):
        lid = row["target_logon_id"]
    else:
        lid = row["subject_logon_id"]
    if lid in (None, "", "0x0", "0X0"):
        return None
    return (host, str(lid).lower())

df["session_key"] = df.apply(assign_session_key, axis=1)
df.head()
````

````python
def build_sessions(df):
    sessions = []
    for sk, g in df.dropna(subset=["session_key"]).groupby("session_key", sort=False):
        g = g.sort_values("ts")
        logon = g[g["event_id"] == 4624].head(1)
        user = (logon["target_user"].iloc[0] if not logon.empty
                else g["subject_user"].dropna().iloc[0] if g["subject_user"].notna().any()
                else "-")
        sessions.append({
            "session_key": sk,
            "host": sk[0],
            "logon_id": sk[1],
            "user": user,
            "logon_type": int(logon["logon_type"].iloc[0]) if not logon.empty and pd.notna(logon["logon_type"].iloc[0]) else None,
            "src_ip": logon["src_ip"].iloc[0] if not logon.empty else None,
            "logon_guid": logon["logon_guid"].iloc[0] if not logon.empty else None,
            "ts_start": g["ts"].min(),
            "ts_end": g["ts"].max(),
            "duration_s": (g["ts"].max() - g["ts"].min()).total_seconds(),
            "n_events": len(g),
            "event_seq": g["event_id"].tolist(),
            "process_seq": [p for p in g["process_name"].fillna("") if p],
        })
    return pd.DataFrame(sessions)

sess_df = build_sessions(df).sort_values("ts_start").reset_index(drop=True)
print(f"Всего сеансов: {len(sess_df)}")
sess_df.head(8)
````

````python
fig, ax = plt.subplots(figsize=(12, max(3, 0.25*len(sess_df))))
colors = {2:"#4caf50", 3:"#1976d2", 7:"#9c27b0", 10:"#ff9800"}
for i, row in sess_df.iterrows():
    c = colors.get(row["logon_type"], "#888")
    ax.barh(i, max((row["ts_end"]-row["ts_start"]).total_seconds()/60, 1),
            left=row["ts_start"], color=c, edgecolor="black", linewidth=0.3)
ax.set_yticks(range(len(sess_df)))
ax.set_yticklabels([f"{r.host}/{r.user[:10]}/{r.logon_type}" for r in sess_df.itertuples()], fontsize=7)
ax.set_xlabel("Время")
ax.set_title("Сеансы по узлам и пользователям (цвет = LogonType)")
plt.tight_layout(); plt.show()
````

## 3. Использование метода префиксных деревьев для поиска подозрительных цепочек событий

**Идея.** Если построить дерево, в которое вложены последовательности событий «нормальных» сеансов, то его узлы запоминают вероятности перехода между событиями. Сеансы, чьи префиксы редки или отсутствуют в дереве, — кандидаты в подозрительные.

Используем последовательность EventID + имя процесса (для 4688) в качестве токенов

### 3.1. Классическое префиксное дерево на строках (демонстрация)

Прежде чем строить префиксное дерево над последовательностями событий, посмотрим на каноническую конструкцию: префиксное дерево (в англоящычной литературе часто называется trie) над набором *строк*. Каждое слово — это путь от корня по символам; общие префиксы разных слов *склеиваются* в одну ветвь, поэтому дерево компактно представляет всё множество.

Реализуем trie на базе готовой библиотеки графов **`networkx.DiGraph`**: каждый узел — это кортеж символов префикса (`()` — корень, `('c',)`, `('c','a')`, `('c','a','r')` …), каждое ребро — добавление одного символа. Узел хранит атрибут `count` — сколько раз этот префикс встретился среди вставленных слов; листовой узел дополнительно помечен флагом `terminal`.


````python
import networkx as nx
import matplotlib.pyplot as plt

def build_string_trie(words):
    """Префиксное дерево на networkx.DiGraph. Узел = кортеж символов префикса."""
    G = nx.DiGraph()
    G.add_node((), count=0, terminal=False)
    for w in words:
        prefix = ()
        for ch in w:
            new_prefix = prefix + (ch,)
            if not G.has_node(new_prefix):
                G.add_node(new_prefix, count=0, terminal=False)
                G.add_edge(prefix, new_prefix, token=ch)
            G.nodes[new_prefix]["count"] += 1
            prefix = new_prefix
        G.nodes[prefix]["terminal"] = True
    return G
````

````python
words = ["car", "card", "care", "cat", "do", "dog", "dome", "domino"]
trie = build_string_trie(words)
````

````python
trie.nodes()
````

````python
def hierarchy_pos(G, root=(), width=1.0, vert_gap=0.3, xcenter=0.5):
    """Простая иерархическая раскладка корневого дерева."""
    pos = {}
    def _layout(node, left, right, depth):
        pos[node] = ((left + right) / 2, -depth * vert_gap)
        children = list(G.successors(node))
        if not children:
            return
        step = (right - left) / len(children)
        for i, ch in enumerate(children):
            _layout(ch, left + i * step, left + (i + 1) * step, depth + 1)
    _layout(root, 0.0, width, 0)
    return pos

# Визуализация
pos = hierarchy_pos(trie, root=(), width=2.0, vert_gap=0.45)
labels = {n: (f"ROOT {trie.nodes[n]['count']}" if n == () else n[-1] + " " + str(trie.nodes[n]['count'])) for n in trie.nodes()}
node_colors = ["#e53935" if trie.nodes[n]["terminal"]
               else ("#888" if n == () else "#ffd180")
               for n in trie.nodes()]

plt.figure(figsize=(11, 4))
nx.draw(trie, pos,
        labels=labels, node_color=node_colors,
        node_size=900, font_size=9, font_weight="bold",
        arrowsize=10, edge_color="#666")
plt.title(f"Префиксное дерево над словами: {words}\n(красные узлы — концы слов)")
plt.tight_layout(); plt.show()
````

### 3.2. От строк к последовательностям событий

Идея переносится один-в-один: вместо символов алфавита у нас будут **токены событий** — `EventID` или, для `4688`, `EventID:имя_процесса`. Сеанс пользователя — это «слово», и набор «нормальных» сеансов формирует trie, по узлам которого можно оценить вероятность каждого перехода и подсветить аномальные.

Для оценки вероятности будем применять меру схожую перплексии - насколько вероятна (или невероятна) вся наблюдаемая последовательность токенов. 
Для последовательности $x_1, x_2, \ldots, x_n$ мера будетт считаться так:

$$\text{NLL}(x) = \frac{1}{n} \sum_{i=1}^{n} -\log P(x_i \mid x_{i-k}, \ldots, x_{i-1})$$

где $k = \texttt{max\_depth} - 1$. То есть мы строим Марковскую модель порядка $k$ — каждый токен «зависит» только от $k$ предшествующих (иначе модель будет крайне чувствительна к любым изменениям в длинных сеансах, что слабо соответствует реальности).

Процесс вычисления вероятности $P(x_i \mid \text{ctx})$:

в дереве узел `ctx` хранит счётчик `count`, равный числу раз, когда n-грамма `ctx` встретилась в обучении. Дополнительно используется сглаживание Лапласа ($\Delta\alpha$):

$$P(x_i \mid \text{ctx}) = \frac{c(\text{ctx}, x_i) + \alpha}{c(\text{ctx}, *) + \alpha \cdot V}$$

где:

- $c(\text{ctx}, x_i)$ — счётчик `count` дочернего узла `ctx + (x_i,)`,
- $c(\text{ctx}, *)$ — сумма счётчиков `count` всех потомков узла `ctx`, то есть сколько раз контекст `ctx` встречался в принципе,
- $V$ — общий размер словаря (`len(self.vocab)`),
- $\alpha$ — параметр сглаживания (`smoothing`, в нашем примере взят равный `0.3`).

Функция сглаживания: избежать ситуации, когда незнакомый переход получал бы меру $\infty$, так как его вероятность равна 0, соответственно логарифм будет равен бесконечности

````python
class PrefixTree:
    """Префиксное дерево с лапласовским сглаживанием 

    Узлы графа — кортежи токенов префиксов; корень — пустой кортеж `()`.
    Каждое ребро — расширение префикса на один токен. Атрибут `count` узла
    равен количеству раз, когда соответствующая n-грамма встречалась при обучении.
    """
    def __init__(self, max_depth=6, smoothing=0.3):
        self.max_depth = max_depth
        self.smoothing = smoothing
        self.G = nx.DiGraph()
        self.G.add_node((), count=0)
        self.vocab = set()

    def fit(self, sequences):
        for seq in sequences:
            for i in range(len(seq)):
                prefix = ()
                for tok in seq[i:i + self.max_depth]:
                    self.vocab.add(tok)
                    new_prefix = prefix + (tok,)
                    if not self.G.has_node(new_prefix):
                        self.G.add_node(new_prefix, count=0)
                        self.G.add_edge(prefix, new_prefix, token=tok)
                    self.G.nodes[new_prefix]["count"] += 1
                    prefix = new_prefix
        return self

    def _ctx_node(self, ctx):
        ctx = tuple(ctx)
        return ctx if self.G.has_node(ctx) else None

    def neg_log_prob(self, seq):
        if not seq:
            return 0.0
        total, cnt = 0.0, 0
        V = max(len(self.vocab), 1)
        for i, tok in enumerate(seq):
            ctx = tuple(seq[max(0, i - self.max_depth + 1):i])
            node = self._ctx_node(ctx)
            if node is not None:
                ctx_total = sum(self.G.nodes[c]["count"] for c in self.G.successors(node))
                child = node + (tok,)
                tok_count = self.G.nodes[child]["count"] if self.G.has_node(child) else 0
                p = (tok_count + self.smoothing) / (ctx_total + self.smoothing * V)
            else:
                p = self.smoothing / (self.smoothing * V)
            total += -math.log(max(p, 1e-12))
            cnt += 1
        return total / cnt


````

````python
def session_to_tokens(row):
    out = []
    proc_seq = list(row["process_seq"]) + ["-"] * len(row["event_seq"])
    for eid, p in zip(row["event_seq"], proc_seq):
        if eid == 4688 and p:
            base = str(p).replace("\\", "/").rsplit("/", 1)[-1].lower()
            out.append(f"4688:{base}")
        else:
            out.append(str(eid))
    return out
````

````python
def anomalous_prefixes(seq, tree, top=5, ctx=3):
    out = []
    V = max(len(tree.vocab), 1)
    for i, tok in enumerate(seq):
        ctx_seq = tuple(seq[max(0, i - ctx):i])
        node = tree._ctx_node(ctx_seq)
        if node is not None:
            ctx_total = sum(tree.G.nodes[c]["count"] for c in tree.G.successors(node))
            child = node + (tok,)
            tok_count = tree.G.nodes[child]["count"] if tree.G.has_node(child) else 0
            p = (tok_count + tree.smoothing) / (ctx_total + tree.smoothing * V)
        else:
            p = tree.smoothing / (tree.smoothing * V)
        out.append((-math.log(max(p, 1e-12)), " → ".join(list(ctx_seq) + [tok])))
    out.sort(reverse=True)
    return out[:top]

worst = sess_df.sort_values("nll", ascending=False).iloc[0]
print(f"Наиболее аномальный сеанс: {worst['host']}/{worst['user']} (NLL={worst['nll']:.2f})")
print("Топ-редкие переходы:")
for nll, path in anomalous_prefixes(worst["tokens"], tree):
    print(f"  -log p = {nll:5.2f}   {path}")
````

````python
def plot_session_tree(seq, max_nodes=25):
    G = nx.DiGraph()
    for i in range(len(seq)-1):
        a, b = seq[i], seq[i+1]
        if a not in G: G.add_node(a)
        if b not in G: G.add_node(b)
        if G.has_edge(a, b):
            G[a][b]["weight"] += 1
        else:
            G.add_edge(a, b, weight=1)
        if len(G.nodes) >= max_nodes: break
    pos = nx.spring_layout(G, seed=1, k=1.2)
    plt.figure(figsize=(10,5))
    nx.draw(G, pos, with_labels=True, node_size=900, node_color="#ffd180",
            font_size=7, arrowsize=12, edge_color="#666")
    plt.title("Граф переходов в подозрительном сеансе"); plt.tight_layout(); plt.show()

plot_session_tree(worst["tokens"])
````

## 4. Выполнение задачи наблюдения за пользователем (Identity tracking): построение расширенного сеанса пользователя за пределами одного узла

В рамках одного хоста сеанс задаётся `(host, logon_id)`. Но реальный пользователь часто переходит между узлами: интерактивно зашёл на WS01, оттуда — сетевой сеанс на FILESERVER, затем remote-interactive на DC01. Чтобы построить **расширенный сеанс**, нужно склеить локальные сеансы по нескольким сигналам.

Сигналы склейки (в порядке убывания надёжности):

1. Совпадение **`LogonGUID`** — Windows проставляет один и тот же GUID для связных логонов в рамках одной Kerberos TGT-сессии.
2. Совпадение `(account, src_ip)` и пересечение по времени.
3. Цепочка: `host_A` устанавливает сетевой сеанс на `host_B` ⇒ `IpAddress` сеанса на `host_B` совпадает с одним из IP-адресов `host_A`.


````python
class DSU:
    def __init__(self, items):
        self.parent = {x: x for x in items}
        self.rank   = {x: 0 for x in items}
    def find(self, x):
        while self.parent[x] != x:
            self.parent[x] = self.parent[self.parent[x]]
            x = self.parent[x]
        return x
    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra == rb: return False
        if self.rank[ra] < self.rank[rb]: ra, rb = rb, ra
        self.parent[rb] = ra
        if self.rank[ra] == self.rank[rb]: self.rank[ra] += 1
        return True

sess_idx = list(sess_df.index)
dsu = DSU(sess_idx)

# Правило A — общий LogonGUID
guid_to_idx = defaultdict(list)
for i, r in sess_df.iterrows():
    if r["logon_guid"] and r["logon_guid"] != "-":
        guid_to_idx[r["logon_guid"]].append(i)
joined_guid = 0
for g, ix in guid_to_idx.items():
    for j in ix[1:]:
        if dsu.union(ix[0], j): joined_guid += 1

# Правило B — (account, src_ip) с временным пересечением ±15 мин
joined_acct = 0
for u, g in sess_df.groupby("user"):
    g = g.sort_values("ts_start")
    rows = list(g.itertuples())
    for i in range(len(rows)):
        for j in range(i+1, len(rows)):
            if rows[j].ts_start - rows[i].ts_end > pd.Timedelta(minutes=15): break
            if rows[i].src_ip and rows[i].src_ip == rows[j].src_ip and rows[i].src_ip != "-":
                if dsu.union(rows[i].Index, rows[j].Index): joined_acct += 1

# Правило C — host_A инициировал сетевой логон на host_B
host_ips = sess_df.groupby("host")["src_ip"].apply(lambda s: set(x for x in s.dropna() if x != "-")).to_dict()
joined_pivot = 0
for j, r in sess_df[sess_df["logon_type"] == 3].iterrows():
    if r["src_ip"] and r["src_ip"] != "-":
        for h, ips in host_ips.items():
            if h == r["host"]: continue
            if r["src_ip"] in ips:
                cand = sess_df[(sess_df["host"] == h) & (sess_df["src_ip"] == r["src_ip"])]
                cand = cand[(cand["ts_start"] <= r["ts_start"]) & (cand["ts_end"] >= r["ts_start"] - pd.Timedelta(minutes=30))]
                if len(cand):
                    if dsu.union(j, cand.index[0]): joined_pivot += 1

print(f"Слияний по LogonGUID: {joined_guid}; по (account,src_ip): {joined_acct}; по pivot: {joined_pivot}")
sess_df["identity_id"] = [dsu.find(i) for i in sess_df.index]
print(f"Расширенных идентичностей: {sess_df['identity_id'].nunique()}; локальных сеансов: {len(sess_df)}")
````

````python
extended = (sess_df.groupby("identity_id")
              .agg(users=("user", lambda s: sorted(set(s))),
                   hosts=("host", lambda s: sorted(set(s))),
                   logon_types=("logon_type", lambda s: sorted({int(x) for x in s if pd.notna(x)})),
                   ts_min=("ts_start","min"), ts_max=("ts_end","max"),
                   n_local=("identity_id","count"))
              .reset_index())
extended["span_s"] = (extended["ts_max"] - extended["ts_min"]).dt.total_seconds()
extended["multi_host"] = extended["hosts"].apply(len) > 1
extended.sort_values(["multi_host","n_local"], ascending=[False, False]).head(10)
````

### 4.3. Identity tracking через Sysmon EventID 1

Sysmon EID 1 (`Process Create`) содержит поле **`LogonGuid`**, идентичное `LogonGuid` соответствующего `4624`. Это даёт **четвёртый сигнал склейки**: если процесс на хосте A был создан с `LogonGuid=g`, и через короткое время на хосте B возникает `4624` с тем же `g` (и `IpAddress` хоста A), процесс — инициатор удалённого сеанса.

Особенно сильный сигнал — когда `Image` процесса является известным инструментом удалённого исполнения: `psexec.exe`, `wmic.exe`, `winrs.exe`, `Invoke-Command`, `PsRemoting`. Доразметим атакующую цепочку Sysmon-событием `psexec` и реализуем правило.

````python
# 4.3.1. Добавим Sysmon EID 1 (process_create) для атакующего: psexec на WS02 → DC01

attacker_sess = sess_df[(sess_df['host'] == 'WS02') & (sess_df['user'] == 'mallory')].iloc[0]
atk_logon_id = attacker_sess['logon_id']
atk_guid     = attacker_sess['logon_guid']

dc_logon = sess_df[(sess_df['host'] == 'DC01') & (sess_df['user'] == 'admin')].iloc[0]
sysmon_ts = dc_logon['ts_start'] - pd.Timedelta(seconds=4)

sysmon_event = {
    '@timestamp': sysmon_ts.isoformat(),
    'Hostname': 'WS02',
    'EventID': 1,
    'Channel': 'Microsoft-Windows-Sysmon/Operational',
    'SubjectUserName': 'mallory',
    'SubjectLogonId': atk_logon_id,
    'LogonGuid': atk_guid,
    'NewProcessName': 'C:\\Tools\\PsExec.exe',
    'CommandLine': 'PsExec.exe \\\\DC01 -u CORP\\admin -p ******** cmd.exe /c whoami',
    'ParentProcessName': 'C:\\Windows\\System32\\cmd.exe',
}
sysmon_event['raw_message'] = (
    'A new process has been created.\n'
    f'Creator Subject:\n  Account Name: mallory\n  Logon ID: {atk_logon_id}\n'
    f'Process Information:\n  New Process Name: {sysmon_event["NewProcessName"]}\n'
    f'  Command Line: {sysmon_event["CommandLine"]}\n  Creator Process Name: {sysmon_event["ParentProcessName"]}'
)

sysmon_norm = normalize_event(sysmon_event)
sysmon_norm['logon_guid'] = atk_guid
sysmon_norm['channel'] = 'Microsoft-Windows-Sysmon/Operational'

base_df = df_v2 if 'df_v2' in dir() else df
if 'member_name' not in base_df.columns:
    base_df = base_df.assign(member_name=None, group_name=None)
sysmon_norm.setdefault('member_name', None); sysmon_norm.setdefault('group_name', None)

df_v3 = pd.concat([base_df, pd.DataFrame([sysmon_norm])], ignore_index=True).sort_values('ts').reset_index(drop=True)
df_v3['session_key'] = df_v3.apply(assign_session_key, axis=1)
print(f'df_v3: {len(df_v3)} событий; добавлен Sysmon-1 process_create на WS02 с LogonGuid={atk_guid}')
````

````python
# 4.3.2. Четвёртое правило склейки: Sysmon-1 (lateral tool) на host_A с LogonGuid=g
#         + 4624 на host_B с тем же LogonGuid в окне 60 сек = pivot-ссылка

REMOTE_TOOLS = re.compile(r'(?i)\b(psexec|wmic|winrs|invoke-command|enter-pssession|impacket|cobalt)\b')

sess_df_v3 = build_sessions(df_v3).sort_values('ts_start').reset_index(drop=True)
sess_df_v3['country'] = sess_df_v3['src_ip'].apply(lambda ip: geoip_lookup(ip)[0] if 'geoip_lookup' in dir() else 'UNKNOWN')
print(f'Локальных сеансов в df_v3: {len(sess_df_v3)}')

dsu_v3 = DSU(list(sess_df_v3.index))

# Базовые правила A/B/C на новом sess_df_v3
guid_to_idx = defaultdict(list)
for i, r in sess_df_v3.iterrows():
    if r['logon_guid'] and r['logon_guid'] != '-':
        guid_to_idx[r['logon_guid']].append(i)
for g, ix in guid_to_idx.items():
    for j in ix[1:]:
        dsu_v3.union(ix[0], j)

# Правило D — Sysmon-1 со lateral-tool инициирует удалённый 4624 с тем же LogonGuid
joined_sysmon = 0
sysmon_lateral = df_v3[(df_v3['event_id'] == 1) &
                       (df_v3['channel'].fillna('').str.contains('Sysmon', case=False)) &
                       (df_v3['cmdline'].fillna('').apply(lambda x: bool(REMOTE_TOOLS.search(str(x)))))]
for _, sm in sysmon_lateral.iterrows():
    if not sm['logon_guid']: continue
    src_host = sm['host']
    cand = df_v3[(df_v3['event_id'] == 4624) & (df_v3['host'] != src_host) &
                 (df_v3['logon_guid'] == sm['logon_guid']) &
                 (df_v3['ts'] >= sm['ts']) & (df_v3['ts'] <= sm['ts'] + pd.Timedelta(seconds=60))]
    for _, c in cand.iterrows():
        src_sk = sm['session_key']; dst_sk = c['session_key']
        if src_sk is None or dst_sk is None: continue
        src_idx = sess_df_v3.index[sess_df_v3['session_key'] == src_sk]
        dst_idx = sess_df_v3.index[sess_df_v3['session_key'] == dst_sk]
        if len(src_idx) and len(dst_idx):
            if dsu_v3.union(int(src_idx[0]), int(dst_idx[0])):
                joined_sysmon += 1

sess_df_v3['identity_id'] = [dsu_v3.find(i) for i in sess_df_v3.index]
print(f'Слияний по Sysmon-1 LogonGuid: {joined_sysmon}')

attacker_id = sess_df_v3[(sess_df_v3['host'] == 'WS02') & (sess_df_v3['user'] == 'mallory')]['identity_id'].iloc[0]
sess_df_v3[sess_df_v3['identity_id'] == attacker_id][['host','user','logon_type','src_ip','logon_guid','ts_start','ts_end']]
````

## 5. Использование обогащения событий (на примере GeoIP) и применение графовых представлений для поиска аномальных цепочек событий

### 5.1. GeoIP

Полноценный MaxMind GeoLite2 требует регистрации, поэтому в практикуме мы используем **встроенный игрушечный GeoIP-резолвер** на основе диапазонов RFC и нескольких публичных подсетей (этого достаточно для иллюстрации идеи; в боевом режиме замените `geoip_lookup` на `geoip2.database.Reader`).

Аномалия GeoIP — это сеанс пользователя `u` с IP, чья страна или ASN не входят в типичный для `u` профиль.

````python
import ipaddress

GEO_TABLE = [
    ("10.0.0.0/8",        "INTERNAL", "CORP-LAN"),
    ("172.16.0.0/12",     "INTERNAL", "CORP-LAN"),
    ("192.168.0.0/16",    "INTERNAL", "CORP-LAN"),
    ("203.0.113.0/24",    "RU",       "AS-EVIL"),
    ("198.51.100.0/24",   "DE",       "AS-PARTNER"),
    ("8.8.8.0/24",        "US",       "GOOGLE"),
    ("1.1.1.0/24",        "US",       "CLOUDFLARE"),
]
GEO_NETS = [(ipaddress.ip_network(n), c, a) for n, c, a in GEO_TABLE]

def geoip_lookup(ip):
    if not ip or ip == "-": return ("UNKNOWN","UNKNOWN")
    try:
        ip_obj = ipaddress.ip_address(ip)
    except ValueError:
        return ("UNKNOWN","UNKNOWN")
    for net, country, asn in GEO_NETS:
        if ip_obj in net:
            return (country, asn)
    return ("FOREIGN","UNKNOWN-ASN")

sess_df[["country","asn"]] = sess_df["src_ip"].apply(lambda x: pd.Series(geoip_lookup(x)))
print("Сеансы по странам/ASN:")
print(sess_df.groupby(["country","asn"]).size().sort_values(ascending=False).to_string())
````

````python
user_country_profile = (sess_df.groupby("user")["country"]
                        .agg(lambda s: Counter(s)).to_dict())

def geo_anomaly(row, top_n=2):
    prof = user_country_profile.get(row["user"], Counter())
    if not prof: return False
    top = {c for c,_ in prof.most_common(top_n)}
    return row["country"] not in top

sess_df["geo_anomaly"] = sess_df.apply(geo_anomaly, axis=1)
geo_alerts = sess_df[sess_df["geo_anomaly"]][["host","user","logon_type","src_ip","country","asn","ts_start"]]
print("Сеансы с GeoIP-аномалией:")
geo_alerts
````

### 5.2. Графовые методы: эмбеддинги и поиск аномальных подграфов

Строим взвешенный гиперграф взаимодействий: вершины — сущности (`user:`, `host:`, `ip:`, `proc:`), рёбра — связи в событиях (логоны, запуски процессов, доступ к файлам). Затем считаем структурные эмбеддинги вершин и применяем `IsolationForest` для поиска вершин, чья соседская структура необычна.

В качестве эмбеддингов используем простой и быстрый набор структурных признаков (степень, кластеризация, разнообразие соседей по типам). Для боевых задач замените на `node2vec` или `GraphSAGE` — формат тот же.

````python
def build_security_graph(df, sess_df):
    G = nx.MultiDiGraph()
    for _, r in sess_df.iterrows():
        u = f"user:{r['user']}"; h = f"host:{r['host']}"
        ip = f"ip:{r['src_ip']}" if r["src_ip"] and r["src_ip"] != "-" else None
        G.add_node(u, kind="user"); G.add_node(h, kind="host")
        G.add_edge(u, h, etype="logon", logon_type=r["logon_type"], country=r["country"])
        if ip:
            G.add_node(ip, kind="ip", country=r["country"], asn=r["asn"])
            G.add_edge(ip, u, etype="from_ip")
    for _, r in df[df["event_id"] == 4688].dropna(subset=["process_name"]).iterrows():
        u = f"user:{r['subject_user']}"
        p = f"proc:{Path(str(r['process_name'])).name.lower()}"
        G.add_node(u, kind="user"); G.add_node(p, kind="proc")
        G.add_edge(u, p, etype="exec")
    return G

G = build_security_graph(df, sess_df)
print(f"Граф: |V|={G.number_of_nodes()}, |E|={G.number_of_edges()}")
kinds = Counter(d.get("kind","?") for _, d in G.nodes(data=True))
print("Типы вершин:", dict(kinds))
````

````python
from sklearn.ensemble import IsolationForest
G_simple = nx.Graph(G)

def node_features(node):
    deg = G_simple.degree(node)
    nb_kinds = Counter(G_simple.nodes[n].get("kind","?") for n in G_simple.neighbors(node))
    cl = nx.clustering(G_simple, node) if deg > 1 else 0.0
    out_logons = sum(1 for _,_,d in G.out_edges(node, data=True) if d.get("etype")=="logon")
    proc_div = sum(1 for n in G_simple.neighbors(node) if G_simple.nodes[n].get("kind")=="proc")
    ip_div = sum(1 for n in G_simple.neighbors(node) if G_simple.nodes[n].get("kind")=="ip")
    return [deg, cl, out_logons, proc_div, ip_div, nb_kinds.get("user",0), nb_kinds.get("host",0)]

nodes = list(G.nodes())
X = np.array([node_features(n) for n in nodes])
iso = IsolationForest(contamination=0.08, random_state=RANDOM_SEED).fit(X)
scores = -iso.score_samples(X)
node_scores = pd.DataFrame({"node": nodes, "kind":[G.nodes[n].get("kind") for n in nodes], "score": scores})
print("Топ-аномальных вершин графа:")
node_scores.sort_values("score", ascending=False).head(10)
````

````python
plt.figure(figsize=(11, 6))
sub_nodes = [n for n in G_simple if G_simple.nodes[n].get("kind") in {"user","host","ip"}]
H = G_simple.subgraph(sub_nodes).copy()
pos = nx.spring_layout(H, seed=2, k=0.7)
score_dict = dict(zip(node_scores["node"], node_scores["score"]))
threshold = np.percentile(scores, 92)
node_colors = []
for n in H.nodes():
    if score_dict.get(n, 0) > threshold:
        node_colors.append("#e53935")
    elif H.nodes[n].get("kind") == "user":
        node_colors.append("#1976d2")
    elif H.nodes[n].get("kind") == "host":
        node_colors.append("#43a047")
    else:
        node_colors.append("#fb8c00")
nx.draw(H, pos, node_color=node_colors, with_labels=True, font_size=7,
        node_size=350, edge_color="#aaaaaa", width=0.6)
plt.title("Граф «пользователь–хост–IP» (красный = аномальная структура)")
plt.tight_layout(); plt.show()
````

### 5.4. Эмбеддинги вершин и поиск аномальных подграфов через MMD

Структурные признаки степени/кластеризации полезны, но плохо передают «контекст» вершины — её характерных соседей через 2–3 хопа. Эмбеддинг-методы (DeepWalk / node2vec / GraphSAGE) решают эту проблему: они представляют вершину низкоразмерным вектором, в котором близкие в графе вершины оказываются близкими в евклидовом смысле.

Здесь мы реализуем **DeepWalk-подобную** схему как pedagogically чистый аналог node2vec ($p=q=1$) и используем эмбеддинги двумя способами:

1. **Поточечная аномалия** — `IsolationForest` на эмбеддингах вершин (заменяет ручные структурные признаки из 5.2).
2. **Anomalous Subgraph Detection** — для каждой вершины-пользователя считаем **MMD** между распределением эмбеддингов её эго-сети (1-hop соседи) и глобальным распределением. Большое MMD — пользователь имеет «нетипичное» окружение.

В production эта же схема воспроизводится на `gensim.models.Word2Vec(sentences=walks, vector_size=64, window=5, sg=1)` или `pyG node2vec`. Здесь ради воспроизводимости и отсутствия дополнительной зависимости — выводим эмбеддинги через PPMI + TruncatedSVD (асимптотически эквивалентно SGNS, см. Levy & Goldberg, 2014).

````python
from sklearn.decomposition import TruncatedSVD

def random_walks(graph, num_walks=12, walk_length=10, seed=RANDOM_SEED):
    rng = random.Random(seed)
    walks = []
    nodes = list(graph.nodes())
    for _ in range(num_walks):
        rng.shuffle(nodes)
        for n in nodes:
            walk = [n]; cur = n
            for _ in range(walk_length - 1):
                neigh = list(graph.neighbors(cur))
                if not neigh: break
                cur = rng.choice(neigh)
                walk.append(cur)
            walks.append(walk)
    return walks

def cooccurrence(walks, window=5):
    co = defaultdict(lambda: defaultdict(int))
    for w in walks:
        for i, tok in enumerate(w):
            for j in range(max(0, i-window), min(len(w), i+window+1)):
                if i == j: continue
                co[tok][w[j]] += 1
    return co

def ppmi_matrix(co, nodes_list):
    idx = {n: i for i, n in enumerate(nodes_list)}
    N = len(nodes_list)
    M = np.zeros((N, N), dtype=np.float32)
    for u, neigh in co.items():
        for v, c in neigh.items():
            if u in idx and v in idx:
                M[idx[u], idx[v]] = c
    rs = M.sum(axis=1, keepdims=True) + 1e-9
    cs = M.sum(axis=0, keepdims=True) + 1e-9
    tot = M.sum() + 1e-9
    PMI = np.log(np.maximum(M * tot / (rs @ cs), 1e-9))
    return np.maximum(PMI, 0.0)

walks = random_walks(G_simple, num_walks=12, walk_length=10)
co = cooccurrence(walks, window=5)
nodes_list_v2 = list(G_simple.nodes())
PPMI = ppmi_matrix(co, nodes_list_v2)
emb = TruncatedSVD(n_components=16, random_state=RANDOM_SEED).fit_transform(PPMI)
emb /= (np.linalg.norm(emb, axis=1, keepdims=True) + 1e-9)
print(f'Эмбеддинги вершин: shape={emb.shape}; первая вершина {nodes_list_v2[0]} -> {emb[0][:5].round(3)}...')
````

````python
# 5.4.1. IsolationForest на эмбеддингах vs на ручных признаках
iso_emb = IsolationForest(contamination=0.08, random_state=RANDOM_SEED).fit(emb)
scores_emb = -iso_emb.score_samples(emb)
node_idx = {n: i for i, n in enumerate(nodes_list_v2)}

top_struct = (node_scores.set_index('node')['score']
              .sort_values(ascending=False).head(8))
top_emb = pd.Series(scores_emb, index=nodes_list_v2).sort_values(ascending=False).head(8)
cmp_methods = pd.concat([top_struct.rename('score_struct'), top_emb.rename('score_emb')], axis=1)
print('Топ-аномалий двух методов (NaN — не входит в топ другого):')
cmp_methods.fillna('-')
````

````python
# 5.4.2. MMD по эго-сетям пользователей

def rbf_kernel(X, Y, sigma=None):
    if sigma is None:
        sigma = float(np.median(np.linalg.norm(X[:,None,:]-X[None,:,:], axis=2)) + 1e-9)
    d = np.linalg.norm(X[:,None,:] - Y[None,:,:], axis=2) ** 2
    return np.exp(-d / (2 * sigma * sigma)), sigma

def mmd2(X, Y, sigma=None):
    '''Несмещённое MMD^2 (Gretton et al. 2012) между двумя выборками эмбеддингов.'''
    if len(X) < 2 or len(Y) < 2:
        return float('nan')
    Kxx, s = rbf_kernel(X, X, sigma)
    Kyy, _ = rbf_kernel(Y, Y, s)
    Kxy, _ = rbf_kernel(X, Y, s)
    n, m = len(X), len(Y)
    sxx = (Kxx.sum() - np.trace(Kxx)) / (n*(n-1))
    syy = (Kyy.sum() - np.trace(Kyy)) / (m*(m-1))
    sxy = Kxy.mean()
    return float(sxx + syy - 2 * sxy)

baseline = emb

mmd_records = []
for n in nodes_list_v2:
    if G_simple.nodes[n].get('kind') != 'user':
        continue
    nbrs = list(G_simple.neighbors(n))
    if len(nbrs) < 3:
        continue
    X = np.stack([emb[node_idx[v]] for v in nbrs if v in node_idx])
    val = mmd2(X, baseline)
    mmd_records.append({'user': n, 'deg': len(nbrs), 'mmd2': val})

mmd_df = pd.DataFrame(mmd_records).sort_values('mmd2', ascending=False)
print('Пользователи, чьё эго-окружение наиболее далеко от глобального распределения (MMD^2):')
mmd_df.head(10)
````

````python
plt.figure(figsize=(8,4))
mmd_df_plot = mmd_df.sort_values('mmd2', ascending=True)
plt.barh(mmd_df_plot['user'].str.replace('user:',''), mmd_df_plot['mmd2'],
         color=['#e53935' if v > mmd_df_plot['mmd2'].quantile(0.85) else '#1976d2' for v in mmd_df_plot['mmd2']])
plt.xlabel('MMD^2 (эго-сеть vs всё)')
plt.title('Аномальность эго-окружения пользователя')
plt.tight_layout(); plt.show()
````

## 6. Формирование и тестирование правил на языке EPL

[Esper](https://esper.espertech.com/) — это open-source CEP-движок с декларативным языком EPL, расширяющим SQL операторами над временными окнами и шаблонами. Боевая интеграция Esper делается из JVM/.NET. Для практикума мы:

* формализуем правила в синтаксисе EPL (как они пойдут в production);
* реализуем **минимальный CEP-движок на Python**, исполняющий аналог этих правил по нашему `DataFrame` с правильной временной семантикой (sliding/length windows + последовательностные паттерны).

Это позволяет тестировать логику правил на исторических данных перед публикацией в Esper.

### 6.1. Правила, которые мы хотим закрыть

1. **Брутфорс**: ≥5 событий `4625` от одного `src_ip` в окне 60 секунд, после чего на тот же хост приходит `4624` с тем же `src_ip`/`target_user` в течение 5 минут.
2. **PowerShell-обфускация**: `4688` со словами `-enc`, `-EncodedCommand`, `FromBase64String` в `cmdline`.
3. **Lateral movement**: успешный `4624` LogonType=3 из IP вне доверенного диапазона (страна ≠ `INTERNAL`) с привилегиями `SeDebugPrivilege` (4672) в том же сеансе.
4. **Ntdsutil/IFM dump**: процесс `ntdsutil.exe` с аргументом `ifm`/`create full`.
5. **Mimikatz-подобные подсказки**: команды `sekurlsa::logonpasswords`, `lsadump::sam`.

````python
ESPER_EPL = r"""
/* ==== Esper EPL правила (production) ==== */

@Name('R1_Bruteforce_then_Success')
insert into Alerts
select 'R1' as rule_id, b.src_ip, s.target_user, s.host
from pattern [
   every b = SecEvent(event_id=4625) ->
   [count(b2 = SecEvent(event_id=4625, src_ip = b.src_ip)) >= 4 within 60 sec] ->
   s = SecEvent(event_id=4624, src_ip = b.src_ip, target_user = b.target_user) within 5 min
];

@Name('R2_PowerShell_Encoded')
insert into Alerts
select 'R2' as rule_id, e.host, e.subject_user, e.cmdline
from SecEvent e
where e.event_id = 4688
  and (e.cmdline regexp '(?i).*-enc(odedCommand)?\b.*'
    or e.cmdline regexp '(?i).*FromBase64String.*');

@Name('R3_Foreign_Net_Logon_with_DebugPriv')
insert into Alerts
select 'R3' as rule_id, l.host, l.target_user, l.src_ip
from pattern [
   every l = SecEvent(event_id=4624, logon_type=3, country != 'INTERNAL') ->
   p = SecEvent(event_id=4672, subject_logon_id = l.target_logon_id,
                privileges like '%SeDebugPrivilege%') within 30 sec
];

@Name('R4_Ntdsutil_IFM_Dump')
insert into Alerts
select 'R4' as rule_id, e.host, e.subject_user, e.cmdline
from SecEvent e
where e.event_id = 4688
  and lower(e.cmdline) like '%ntdsutil%'
  and (lower(e.cmdline) like '%ifm%' or lower(e.cmdline) like '%create full%');

@Name('R5_Mimikatz_keywords')
insert into Alerts
select 'R5' as rule_id, e.host, e.subject_user, e.cmdline
from SecEvent e
where e.event_id = 4688
  and (lower(e.cmdline) like '%sekurlsa::%'
    or lower(e.cmdline) like '%lsadump::%');
"""

print(ESPER_EPL[:600] + "...\n")
print(f"Длина EPL-конфигурации: {len(ESPER_EPL)} символов")
````

````python
# Мини-CEP-движок на Python, повторяющий семантику правил выше.

def run_rules(df, sess_df):
    df = df.copy().sort_values("ts").reset_index(drop=True)
    alerts = []

    # R1: brute-force then success
    fails = df[df["event_id"] == 4625][["ts","src_ip","target_user","host"]].dropna(subset=["src_ip"])
    succ  = df[df["event_id"] == 4624][["ts","src_ip","target_user","host","target_logon_id"]].dropna(subset=["src_ip"])
    for ip, gf in fails.groupby("src_ip"):
        gf = gf.sort_values("ts")
        ts = gf["ts"].tolist()
        for i in range(len(ts)):
            j = i
            while j+1 < len(ts) and (ts[j+1]-ts[i]).total_seconds() <= 60:
                j += 1
            if j-i+1 >= 5:
                window_end = ts[j] + pd.Timedelta(minutes=5)
                cand = succ[(succ["src_ip"] == ip) & (succ["ts"] >= ts[j]) & (succ["ts"] <= window_end)]
                for _, s in cand.iterrows():
                    alerts.append({"rule":"R1","ts":s["ts"],"host":s["host"],
                                   "user":s["target_user"],"detail":f"5+ failures from {ip}"})
                break

    # R2: powershell encoded
    pat = re.compile(r"(?i)(\B-enc(?:odedCommand)?\b|FromBase64String)")
    for _, r in df[df["event_id"] == 4688].dropna(subset=["cmdline"]).iterrows():
        if pat.search(str(r["cmdline"])):
            alerts.append({"rule":"R2","ts":r["ts"],"host":r["host"],
                           "user":r["subject_user"],"detail":str(r["cmdline"])[:120]})

    # R3: foreign Type=3 logon + 4672 SeDebugPrivilege
    foreign = df[(df["event_id"] == 4624) & (df["logon_type"] == 3) & (df["country"] != "INTERNAL")]
    for _, l in foreign.iterrows():
        priv = df[(df["event_id"] == 4672) &
                  (df["subject_logon_id"] == l["target_logon_id"]) &
                  (df["ts"] >= l["ts"]) & (df["ts"] <= l["ts"] + pd.Timedelta(seconds=30)) &
                  (df["privileges"].fillna("").str.contains("SeDebugPrivilege"))]
        if len(priv):
            alerts.append({"rule":"R3","ts":l["ts"],"host":l["host"],"user":l["target_user"],
                           "detail":f"Type=3 from {l['src_ip']} ({l['country']}) + DebugPriv"})

    # R4: ntdsutil ifm
    for _, r in df[df["event_id"] == 4688].dropna(subset=["cmdline"]).iterrows():
        cl = str(r["cmdline"]).lower()
        if "ntdsutil" in cl and ("ifm" in cl or "create full" in cl):
            alerts.append({"rule":"R4","ts":r["ts"],"host":r["host"],
                           "user":r["subject_user"],"detail":str(r["cmdline"])[:120]})

    # R5: mimikatz / lsadump
    for _, r in df[df["event_id"] == 4688].dropna(subset=["cmdline"]).iterrows():
        cl = str(r["cmdline"]).lower()
        if "sekurlsa::" in cl or "lsadump::" in cl:
            alerts.append({"rule":"R5","ts":r["ts"],"host":r["host"],
                           "user":r["subject_user"],"detail":str(r["cmdline"])[:120]})

    return pd.DataFrame(alerts).sort_values("ts").reset_index(drop=True) if alerts else pd.DataFrame(columns=["rule","ts","host","user","detail"])

# Подмешаем country в df для R3
host_country = sess_df.set_index("session_key")["country"].to_dict()
df["country"] = df["session_key"].map(lambda k: host_country.get(k, "UNKNOWN")).fillna("UNKNOWN")

alerts = run_rules(df, sess_df)
print(f"Сработавших алертов: {len(alerts)}")
alerts
````

### 6.2. Проверка качества правил

Сравним результат правил с ground truth — сеансом атакующего, чтобы оценить precision/recall на нашем датасете.

В синтетическом источнике атакующий — `mallory`/`admin` (после захвата). В реальных Mordor-датасетах — описание атакующего находится в YAML-метаданных рядом с zip-файлом. Здесь мы возьмём упрощённую разметку: сеанс «плохой», если включает `mimikatz`, `ntdsutil`, или последовавший за brute-force успех.

````python
def label_session_truth(row):
    cl = " ".join(p.lower() for p in (row["process_seq"] or []) if p)
    return any(k in cl for k in ["mimikatz", "ntdsutil"])

sess_df["truth_bad"] = sess_df.apply(label_session_truth, axis=1)

alerts_keys = set()
for _, a in alerts.iterrows():
    cand = sess_df[(sess_df["host"]==a["host"]) & (sess_df["ts_start"] <= a["ts"]) & (sess_df["ts_end"] >= a["ts"])]
    for k in cand["session_key"]:
        alerts_keys.add(k)
sess_df["detected"] = sess_df["session_key"].isin(alerts_keys)

tp = ((sess_df["truth_bad"]) & (sess_df["detected"])).sum()
fp = ((~sess_df["truth_bad"]) & (sess_df["detected"])).sum()
fn = ((sess_df["truth_bad"]) & (~sess_df["detected"])).sum()
tn = ((~sess_df["truth_bad"]) & (~sess_df["detected"])).sum()
prec = tp/max(tp+fp,1); rec = tp/max(tp+fn,1)
print(f"TP={tp}  FP={fp}  FN={fn}  TN={tn}")
print(f"Precision={prec:.2f}  Recall={rec:.2f}")
sess_df[sess_df["truth_bad"] | sess_df["detected"]][["host","user","logon_type","truth_bad","detected"]]
````

5. *Длинные временные окна.* Перейти от поточных правил Esper к гибридному пайплайну (CEP — быстрые сигналы, Spark/Flink — агрегации за сутки/недели). Что попадёт в каждый слой?
