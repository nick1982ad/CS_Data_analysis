### Парсинг разметки

В `combined_windows.json` для каждого файла указан список интервалов
(каждый — пара `[t_start, t_end]`), в которых произошла аномалия.
Преобразуем оконную разметку в бинарный вектор той же длины, что и ряд
(`1` — точка попадает в окно аномалии, иначе `0`).

```python
LABELS_FILE
```

```python
def load_series_with_labels(csv_path, labels_path, file_key):
    df = pd.read_csv(csv_path, parse_dates=["timestamp"]).sort_values("timestamp").reset_index(drop=True)
    df["label"] = 0
    with open(labels_path, "r", encoding="utf-8") as f:
        labels = json.load(f)
    windows = labels.get(file_key, [])
    for w_start, w_end in windows:
        w_start = pd.to_datetime(w_start) 
        w_end = pd.to_datetime(w_end)
        mask = (df["timestamp"] >= w_start) & (df["timestamp"] <= w_end)
        df.loc[mask, "label"] = 1
    return df, windows

df, anomaly_windows = load_series_with_labels(PRIMARY_FILE, LABELS_FILE, FILE_KEY)
print("Размер ряда:", len(df))
print("Период:", df["timestamp"].min(), "→", df["timestamp"].max())
print("Доля точек, помеченных как аномалия:", round(df["label"].mean() * 100, 2), "%")
print("Окна аномалий:")
for w in anomaly_windows:
    print(" ", w)
df.head()
```

### Интерактивная визуализация ряда (Plotly)

Plotly удобно использовать в задачах мониторинга: можно зумировать на
интересные участки, наводить курсор для просмотра значений, скрывать
отдельные серии. Здесь мы накладываем разметку аномалий как прямоугольные

```python
def plot_series_plotly(df, anomaly_windows, title="Временной ряд",
                       value_col="value"):
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df["timestamp"], y=df[value_col],
        mode="lines", name=value_col,
        line=dict(color="steelblue", width=1)
    ))
    for w_start, w_end in anomaly_windows:
        fig.add_vrect(x0=pd.to_datetime(w_start), x1=pd.to_datetime(w_end),
                      fillcolor="red", opacity=0.15, line_width=0,
                      annotation_text="GT", annotation_position="top left")
    fig.update_layout(title=title, height=420, xaxis_title="Время",
                      yaxis_title=value_col, hovermode="x unified",
                      template="plotly_white")
    return fig

fig = plot_series_plotly(df, anomaly_windows, title="EC2 CPU utilization — интерактивный график")
fig.show()
```

## 3. Предобработка временного ряда

### Стационарность. Тест Дики-Фуллера (ADF)

Многие классические методы (ARIMA, статистические пороги по среднему
и дисперсии) предполагают, что свойства ряда — математическое ожидание,
дисперсия, автокорреляция — не зависят от времени, т.е. ряд **стационарен**.

ADF (Augmented Dickey-Fuller test) - улучшенный тест Дики-Фуллера - это статистический тест на наличие единичного корня в авторегрессионной модели ряда. 

Основная идея метода: если ряд можно описать как $y_t = \rho y_{t-1} + \ldots + \varepsilon_t$, то поведение ряда определяется значением $\rho$:

- $|\rho| < 1$ — ряд **стационарен**: возмущения затухают, ряд возвращается к своему среднему
- $\rho = 1$ — у уравнения есть **единичный корень**: возмущения "запоминаются", ряд блуждает, дисперсия растёт со временем
- $|\rho| > 1$ — взрывной процесс (на практике встречается редко)

ADF тестирует гипотезу $\rho = 1$ против $\rho < 1$.

Тест не оценивает $\rho$ напрямую — для удобства уравнение переписывают через разности первого порядка. Подставляя $\gamma = \rho - 1$, получаем тестовую регрессию:

$$\Delta y_t = \alpha + \beta t + \gamma\, y_{t-1} + \sum_{i=1}^{p} \delta_i\, \Delta y_{t-i} + \varepsilon_t$$

где $\Delta y_t = y_t - y_{t-1}$. Тогда:

- $H_0: \gamma = 0$ ⇔ $\rho = 1$ — есть единичный корень, ряд нестационарен.
- $H_1: \gamma < 0$ ⇔ $\rho < 1$ — единичного корня нет, ряд стационарен.

Добавки $\sum \delta_i \Delta y_{t-i}$ (от этого «Augmented» в названии) убирают автокорреляцию в остатках $\varepsilon_t$ — без них тест работает корректно только для строго AR(1)-процессов. Параметр `autolag="AIC"` в `statsmodels` подбирает оптимальное число лагов $p$.

#### Интерпретация результатов теста
ADF-статистика - это t-статистика значения $\gamma$. Так как мы проверяем гипотезу, что $\gamma = 0$, то чем дальше ADF-статистика от 0, тем вероятнее, что $H_0$ не подтверждается.

p-value в ADF — это вероятность получить ADF-статистику **не больше** (т.е. ещё более отрицательную или такую же), как наблюдаемая, **при условии, что $H_0$ верна** (ряд действительно нестационарный):

$$\text{p-value} = P(t_\gamma \leq t_{\text{набл}} \mid H_0)$$

Так как тест односторонний с альтернативой «$\gamma < 0$», маленькое p-value означает, что наблюдаемая статистика лежит глубоко в левом хвосте, что маловероятно при $H_0$ → отвергаем $H_0$ → ряд можно считать стационарным.

**Правило принятия решения** (типовое):

- p-value $< 0.01$ — сильное свидетельство стационарности
- p-value $< 0.05$ — обычно отвергаем $H_0$
- p-value $< 0.10$ — слабое свидетельство, на границе
- p-value $\geq 0.10$ — отвергнуть $H_0$ нельзя, ряд считается нестационарным

Критические значения 1%, 5%, 10%

Это **квантили распределения Дики-Фуллера** для текущего размера выборки и спецификации модели. Они задают пороги отвержения $H_0$ при выбранном уровне значимости $\alpha$:

- **Критическое значение 1%** — порог, ниже которого статистика отвергает $H_0$ с уровнем значимости 1%.
- **Критическое значение 5%** — то же при 5%.
- **Критическое значение 10%** — то же при 10%.

**Эквивалентный решающий критерий** (в сравнении с p-value) — сравнить наблюдаемую ADF-статистику с этими порогами:

- $t_{\text{набл}} < t_{1\%}$ — отвергаем $H_0$ на уровне 1%, очень сильное свидетельство стационарности.
- $t_{1\%} \leq t_{\text{набл}} < t_{5\%}$ — отвергаем на 5%
- $t_{5\%} \leq t_{\text{набл}} < t_{10\%}$ — отвергаем на 10%
- $t_{\text{набл}} \geq t_{10\%}$ — не отвергаем $H_0$ ни на каком разумном уровне


#### Дополнительные практические нюансы

**Низкая мощность теста.** ADF знаменит тем, что часто не подтвержжает стационарность, даже когда она есть, особенно на коротких рядах и при $\rho$, близком к 1. Принятие $H_0$ — это **не подтверждение** нестационарности, а просто отсутствие достаточных свидетельств в пользу стационарности.

**Дополняющий тест KPSS** имеет инвертированную нулевую гипотезу ($H_0$: ряд стационарен). Если оба теста согласуются, то вывод надёжен. Если расходятся — нужен дополнительный анализ.

**Структурные сдвиги ADF не учитывает.** Если в ряду есть резкий сдвиг, ADF может ложно классифицировать стационарный ряд как нестационарный

**Сезонность.** ADF не вычитает сезонную компоненту автоматически. На рядах с сильной сезонностью имеет смысл сначала декомпозировать (например, через STL) и тестировать остатки отдельно.

```python
def adf_report(x):
    res = adfuller(x.dropna(), autolag="AIC")
    print(f"ADF statistic: {res[0]:.4f}")
    print(f"p-value:       {res[1]:.4f}")
    for k, v in res[4].items():
        print(f"  critical value {k}: {v:.4f}")
    if res[1] < 0.05:
        print("→ Ряд можно считать стационарным (отвергаем H0).")
    else:
        print("→ Нельзя отвергнуть H0 — ряд скорее всего нестационарен.")

adf_report(df["value"])
```

### Декомпозиция STL

**Идея.** STL (Seasonal-Trend decomposition using Loess) представляет ряд как
сумму трёх компонент:

$$y_t = T_t + S_t + R_t$$

где $T_t$ — тренд (медленная составляющая), $S_t$ — сезонность,
$R_t$ — остаток. Сглаживание производится локальной регрессией (LOESS).

**Зачем в мониторинге.** Позволяет отделить «нормальную» периодичность
(например, суточные циклы нагрузки) от подозрительных отклонений. Большие
остатки $R_t$ — кандидаты в аномалии (см. метод 4.5 ниже).

**Ограничения.** Требует разумного периода сезонности (`period`),
плохо работает при множественной/нестационарной сезонности и резких
структурных сдвигах. Аддитивная модель не подходит для рядов, где амплитуда
сезонности пропорциональна уровню (там нужно логарифмирование или
мультипликативная декомпозиция).

```python
step_minutes = int(deltas.median().total_seconds() // 60)
period = max(2, int(24 * 60 / step_minutes))
print(f"Шаг: {step_minutes} мин → период (1 сутки): {period} точек")

stl = STL(df["value"].values, period=period, robust=True).fit()
fig, axs = plt.subplots(4, 1, figsize=(14, 8), sharex=True)
axs[0].plot(df["timestamp"], df["value"], color="steelblue"); axs[0].set_title("Исходный ряд")
axs[1].plot(df["timestamp"], stl.trend, color="darkorange"); axs[1].set_title("Тренд")
axs[2].plot(df["timestamp"], stl.seasonal, color="seagreen"); axs[2].set_title("Сезонность")
axs[3].plot(df["timestamp"], stl.resid, color="crimson"); axs[3].set_title("Остаток")
plt.tight_layout(); plt.show()

df["stl_resid"] = stl.resid
df["stl_trend"] = stl.trend
df["stl_seasonal"] = stl.seasonal
```

## 4. Поиск аномалий

Будем сравнивать методы по одной и той же метрике. Аномальные **окна**
NAB уже преобразованы в точечную бинарную разметку `df["label"]`. Для каждого
метода получим предсказание $\hat y \in \{0, 1\}$ и посчитаем precision/recall/F1.

### Утилиты для оценки и сравнения

```python
results_ad = []  # сюда копим результаты всех методов

def eval_predictions(y_true, y_pred, method_name, verbose=True):
    p = precision_score(y_true, y_pred, zero_division=0)
    r = recall_score(y_true, y_pred, zero_division=0)
    f = f1_score(y_true, y_pred, zero_division=0)
    n_pred = int(np.sum(y_pred))
    if verbose:
        print(f"{method_name:30s} | precision={p:.3f}  recall={r:.3f}  F1={f:.3f}  | предсказано аномалий: {n_pred}")
    results_ad.append({"method": method_name, "precision": p, "recall": r, "F1": f, "n_pred": n_pred})
    return p, r, f

def plot_predictions(df, pred_col, title):
    fig, ax = plt.subplots(figsize=(14, 3.5))
    ax.plot(df["timestamp"], df["value"], lw=0.7, color="steelblue", label="value")
    for w_start, w_end in anomaly_windows:
        ax.axvspan(pd.to_datetime(w_start), pd.to_datetime(w_end),
                   color="red", alpha=0.15)
    anom = df[df[pred_col] == 1]
    ax.scatter(anom["timestamp"], anom["value"], color="red", s=14, zorder=5,
               label="Предсказанная аномалия")
    ax.set_title(title); ax.legend(loc="upper right")
    plt.tight_layout(); plt.show()
```

### 4.1 Z-score (глобальный)

**Идея.** Считаем стандартизированное отклонение каждой точки от глобального среднего:

$$z_t = \frac{y_t - \mu}{\sigma}, \quad \text{аномалия, если } |z_t| > k$$

где $\mu, \sigma$ — выборочные среднее и стандартное отклонение по всему ряду,
$k$ обычно 2.5–3.5.

**Когда работает.** Стационарные ряды без сильной сезонности; нужен быстрый
baseline.

**Ограничения.** $\mu, \sigma$ сами портятся аномалиями (нестойкость к выбросам);
не учитывают временную структуру и тренд; на нестационарных данных дают много
ложных срабатываний.

```python
def detect_zscore(values, k=3.0):
    mu, sigma = np.mean(values), np.std(values)
    z = (values - mu) / sigma
    return (np.abs(z) > k).astype(int)

df["pred_zscore"] = detect_zscore(df["value"].values, k=3.0)
eval_predictions(df["label"], df["pred_zscore"], "Z-score (k=3)")
plot_predictions(df, "pred_zscore", "Z-score (k=3)")
```

### 4.2 Модифицированный Z-score (на основе MAD)

**Идея.** Заменяем неустойчивые среднее и std на устойчивые статистики —
медиану и медианное абсолютное отклонение (MAD):

$$\text{MAD} = \text{median}(|y_t - \text{median}(y)|), \quad
M_t = \frac{0.6745 \cdot (y_t - \text{median}(y))}{\text{MAD}}$$

Точка считается аномалией, если $|M_t| > 3.5$ (классический порог Iglewicz & Hoaglin).

**Преимущество.** MAD устойчив к выбросам (точка пробивания ~50%), поэтому
параметры порога не «портятся» самими аномалиями.

**Ограничения.** По-прежнему игнорирует время/сезонность; на длинных рядах
с трендом будет ловить целые «полки», а не точечные выбросы.

```python
def detect_mad(values, k=3.5):
    med = np.median(values)
    mad = np.median(np.abs(values - med)) + 1e-9
    M = 0.6745 * (values - med) / mad
    return (np.abs(M) > k).astype(int)

df["pred_mad"] = detect_mad(df["value"].values, k=3.5)
eval_predictions(df["label"], df["pred_mad"], "Modified Z-score (MAD)")
plot_predictions(df, "pred_mad", "Modified Z-score / MAD (k=3.5)")
```

### 4.3 Метод межквартильного размаха (IQR)

**Идея.** Аномалии определяются как точки за «усами» бокс-плота:

$$Q_1 = P_{25}, \quad Q_3 = P_{75}, \quad \text{IQR} = Q_3 - Q_1$$
$$y_t < Q_1 - \alpha \cdot \text{IQR} \;\; \text{или} \;\; y_t > Q_3 + \alpha \cdot \text{IQR}$$

Классический множитель $\alpha = 1.5$ (далёкие выбросы) или $3.0$ (extreme).

**Преимущество.** Простой непараметрический метод; не требует
предположения о нормальности.

**Ограничения.** Та же проблема — не учитывает временную структуру; на
многомодальных распределениях интерквартильный размах может быть малоинформативен.

```python
def detect_iqr(values, alpha=1.5):
    q1, q3 = np.percentile(values, [25, 75])
    iqr = q3 - q1
    low, high = q1 - alpha * iqr, q3 + alpha * iqr
    return ((values < low) | (values > high)).astype(int)

df["pred_iqr"] = detect_iqr(df["value"].values, alpha=3.0)
eval_predictions(df["label"], df["pred_iqr"], "IQR (alpha=3.0)")
plot_predictions(df, "pred_iqr", "IQR (alpha=3.0)")
```

### 4.4 Скользящее правило 3σ

**Идея.** Считаем среднее и std в скользящем окне ширины $w$ и сравниваем
текущую точку с локальной нормой:

$$\mu_t = \text{mean}(y_{t-w+1..t}), \quad \sigma_t = \text{std}(y_{t-w+1..t})$$
$$\text{аномалия, если } |y_t - \mu_t| > k \cdot \sigma_t$$

**Преимущество.** Адаптируется к медленно меняющемуся среднему уровню,
ловит локальные «всплески».

**Ограничения.** Размер окна — гиперпараметр; в самом окне могут быть
аномалии, что искажает $\mu, \sigma$ (полезно использовать робастные
варианты — медиану и MAD по окну).

```python
def detect_rolling_sigma(series, window=288, k=3.0):
    s = pd.Series(series)
    mu = s.rolling(window, min_periods=window//2).mean()
    sigma = s.rolling(window, min_periods=window//2).std()
    return ((s - mu).abs() > k * sigma).fillna(False).astype(int).values

df["pred_rolling3s"] = detect_rolling_sigma(df["value"].values, window=period, k=3.0)
eval_predictions(df["label"], df["pred_rolling3s"], f"Rolling 3-sigma (w={period})")
plot_predictions(df, "pred_rolling3s", f"Rolling 3-sigma, window={period} (1 сутки)")
```

### 4.5 Аномалии по остаткам STL

**Идея.** После декомпозиции $y_t = T_t + S_t + R_t$ остаток $R_t$ должен быть
«белым шумом». Применяем к нему любое правило-порог:

$$\text{аномалия, если } |R_t| > k \cdot \sigma_R$$

Часто $\sigma_R$ оценивают как $\text{MAD}(R_t) / 0.6745$ (устойчиво).

**Преимущество.** Явно убирает тренд и сезонность — методы по уровню перестают
давать ложные срабатывания на периодических пиках.

**Ограничения.** Качество зависит от корректного определения периода
сезонности и от устойчивости STL к самим аномалиям (используем `robust=True`).

```python
def detect_stl_residual(residuals, k=3.0):
    r = pd.Series(residuals)
    mad = np.median(np.abs(r - np.median(r))) + 1e-9
    sigma_robust = mad / 0.6745
    return (np.abs(r - np.median(r)) > k * sigma_robust).astype(int).values

df["pred_stl"] = detect_stl_residual(df["stl_resid"].values, k=4.0)
eval_predictions(df["label"], df["pred_stl"], "STL residual + robust 3sigma")
plot_predictions(df, "pred_stl", "Аномалии по остаткам STL (robust)")
```

### 4.6 Isolation Forest

**Идея.** Строится ансамбль случайных деревьев. На каждом шаге выбирается
случайный признак и случайный порог; листовая глубина $h(x)$ для аномальной
точки в среднем оказывается **меньше**, чем для нормальной (аномалии легче
«отрезать»). Аномальный скор:

$$s(x, n) = 2^{-\frac{E[h(x)]}{c(n)}}, \quad c(n) \approx 2 H(n-1) - \frac{2(n-1)}{n}$$

где $H$ — гармоническое число. $s \to 1$ — аномалия, $s \to 0.5$ — норма.

**Сильные стороны.** Не требует разметки; масштабируется на многомерные данные;
быстрая и популярная в продакшене.

**Ограничения.** Чувствителен к параметру `contamination` (доля аномалий
в данных); плохо работает с явно периодическими данными без признаков
(нужно подавать lag-фичи или остатки).

```python
# Подаём признаки: значение, остаток STL, скользящие статистики
feat = pd.DataFrame({
    "value": df["value"].values,
    "stl_resid": df["stl_resid"].values,
    "roll_mean": pd.Series(df["value"]).rolling(period, min_periods=1).mean().values,
    "roll_std":  pd.Series(df["value"]).rolling(period, min_periods=1).std().fillna(0).values,
})

iso = IsolationForest(contamination=0.02, random_state=42, n_estimators=200)
df["pred_iforest"] = (iso.fit_predict(feat) == -1).astype(int)
eval_predictions(df["label"], df["pred_iforest"], "Isolation Forest")
plot_predictions(df, "pred_iforest", "Isolation Forest (contamination=0.02)")
```

### 4.7 Local Outlier Factor (LOF)

**Идея.** Сравнивает локальную плотность точки с плотностью её
$k$-ближайших соседей. Точка-аномалия имеет существенно более низкую
плотность, чем её соседи:

$$\text{LOF}_k(p) = \frac{\sum_{o \in N_k(p)} \frac{\text{lrd}_k(o)}{\text{lrd}_k(p)}}{|N_k(p)|}$$

$\text{LOF} \approx 1$ — норма, $\text{LOF} \gg 1$ — аномалия.

**Преимущество.** Обнаруживает локальные аномалии (когда соседняя область
плотная, а точка вылетает) — полезно в неоднородных данных.

**Ограничения.** $O(n^2)$ по времени без специальных индексов; требует выбора $k$;
плохо работает на очень высоких размерностях (curse of dimensionality).

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.02)
df["pred_lof"] = (lof.fit_predict(feat) == -1).astype(int)
eval_predictions(df["label"], df["pred_lof"], "Local Outlier Factor")
plot_predictions(df, "pred_lof", "LOF (n_neighbors=20)")
```

### 4.8 One-Class SVM

**Идея.** Строит границу, отделяющую основную массу нормальных точек от
начала координат в спрямлённом ядром пространстве:

$$\min_{w, \rho, \xi} \frac{1}{2}\|w\|^2 + \frac{1}{\nu n}\sum_i \xi_i - \rho$$

Параметр $\nu \in (0, 1]$ — верхняя граница доли аномалий и нижняя граница
доли опорных векторов.

**Преимущество.** Гибкость за счёт ядер (RBF и др.); хорошо работает на
небольших данных без явной разметки.

**Ограничения.** Очень чувствителен к масштабированию признаков;
$O(n^2)$ по времени; чувствителен к подбору $\gamma$ и $\nu$.

```python
scaler = StandardScaler()
feat_scaled = scaler.fit_transform(feat)

ocsvm = OneClassSVM(kernel="rbf", gamma="scale", nu=0.02)
ocsvm.fit(feat_scaled)
df["pred_ocsvm"] = (ocsvm.predict(feat_scaled) == -1).astype(int)
eval_predictions(df["label"], df["pred_ocsvm"], "One-Class SVM (RBF)")
plot_predictions(df, "pred_ocsvm", "One-Class SVM (RBF, nu=0.02)")
```

### 4.9 SARIMA-based: prediction interval

**Идея.** Обучаем модель **SARIMA** на нормальной части ряда. Для каждой точки
получаем одношаговый прогноз $\hat y_t$ и доверительный интервал
$[\hat y_t - z_\alpha \sigma_t, \hat y_t + z_\alpha \sigma_t]$. Точка
объявляется аномалией, если фактическое значение **выходит за интервал**.

Формула SARIMA$(p,d,q)(P,D,Q)_s$:

$$\Phi_P(L^s)\, \phi_p(L)\, (1-L)^d (1-L^s)^D y_t = \Theta_Q(L^s)\, \theta_q(L)\, \varepsilon_t$$

**Преимущество.** Учитывает тренд + сезонность; даёт интерпретируемую полосу
«нормального коридора».

**Ограничения.** Требует подбора параметров; долгая инференс-фаза на длинных
рядах; плохо реагирует на структурные сдвиги (после break нужен
переподгон модели).

```python
# обучаем на первой половине ряда (предположим — там «норма»),
# затем используем модель для всего ряда (one-step-ahead in-sample residuals)
train_len = int(len(df) * 0.5)
y = df["value"].values

# минимальная SARIMA: (1,1,1)x(1,1,1, period). Здесь period может быть большой —
# ограничим сезонность 24 точками для скорости (~2 часа при 5-минутном шаге).
seasonal_period = min(period, 24)
try:
    sarima_model = SARIMAX(y[:train_len], order=(1,1,1), seasonal_order=(1,1,1, seasonal_period),
                           enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)
    # one-step-ahead прогнозы по всему ряду
    pred_obj = sarima_model.get_prediction(start=0, end=len(y)-1, dynamic=False)
    mean = pred_obj.predicted_mean
    ci = pred_obj.conf_int(alpha=0.001)  # 99.9%
    lo, hi = ci[:, 0], ci[:, 1]
    df["sarima_mean"] = mean; df["sarima_lo"] = lo; df["sarima_hi"] = hi
    df["pred_sarima"] = ((y < lo) | (y > hi)).astype(int)
    eval_predictions(df["label"], df["pred_sarima"], "SARIMA prediction interval")
    plot_predictions(df, "pred_sarima", "SARIMA-based AD (99.9% interval)")
except Exception as e:
    print("SARIMA не построилась:", e)
```

### 4.10 Prophet-based: prediction interval

**Идея.** Аналогично SARIMA, но используем модель Prophet, которая
аддитивно раскладывает ряд на тренд (кусочно-линейный или логистический),
несколько сезонностей и эффекты праздников:

$$y_t = g(t) + s(t) + h(t) + \varepsilon_t$$

Prophet даёт `yhat_lower` и `yhat_upper`; точка считается аномалией, если выходит
за этот интервал.

**Преимущество.** Робастна к пропускам и выбросам; удобна для бизнес-данных
с явной сезонностью; почти не требует тюнинга.

**Ограничения.** Хуже работает на быстро меняющихся системных метриках с
плохо выраженной периодичностью; «коробочный» интервал может быть слишком широким.

```python
pdf = df[["timestamp", "value"]].rename(columns={"timestamp": "ds", "value": "y"})
m = Prophet(daily_seasonality=True, weekly_seasonality=False,
            yearly_seasonality=False, interval_width=0.999)
m.fit(pdf.iloc[:train_len])
fc = m.predict(pdf[["ds"]])
df["prophet_yhat"] = fc["yhat"].values
df["prophet_lo"] = fc["yhat_lower"].values
df["prophet_hi"] = fc["yhat_upper"].values
df["pred_prophet"] = ((df["value"] < df["prophet_lo"]) |
                      (df["value"] > df["prophet_hi"])).astype(int)
eval_predictions(df["label"], df["pred_prophet"], "Prophet prediction interval")
plot_predictions(df, "pred_prophet", "Prophet-based AD (99.9% interval)")
```

### 4.11 LSTM-Autoencoder

**Идея.** LSTM-Autoencoder учит сжимать последовательности фиксированной длины
$w$ в латентный вектор и затем восстанавливать их. Модель обучается на
«нормальных» окнах; ошибка восстановления

$$\text{err}_t = \frac{1}{w}\sum_{i=0}^{w-1}\left(y_{t-i} - \hat y_{t-i}\right)^2$$

для аномальных окон оказывается значительно выше — отсюда пороговое правило.

**Преимущество.** Захватывает сложные нелинейные временные паттерны;
работает «без учителя» (нужны только нормальные окна).

**Ограничения.** Требует много данных и аккуратного выбора $w$; чувствителен
к гиперпараметрам; сложнее в продакшене (нужно поддерживать модель и порог,
отслеживать concept drift).

```python
from tensorflow.keras import layers, models
tf.random.set_seed(42)

w = 30  # длина окна
series = df["value"].values.astype("float32")
mu_, sd_ = series[:train_len].mean(), series[:train_len].std() + 1e-9
norm = (series - mu_) / sd_

def make_windows(x, w):
    return np.stack([x[i:i+w] for i in range(len(x) - w + 1)])

X_all = make_windows(norm, w)
X_train = X_all[:train_len - w + 1]

inp = layers.Input(shape=(w, 1))
x = layers.LSTM(32, activation="tanh", return_sequences=False)(inp)
x = layers.RepeatVector(w)(x)
x = layers.LSTM(32, activation="tanh", return_sequences=True)(x)
out = layers.TimeDistributed(layers.Dense(1))(x)
ae = models.Model(inp, out)
ae.compile(optimizer="adam", loss="mse")
ae.fit(X_train[..., None], X_train[..., None], epochs=10,
       batch_size=64, verbose=0, shuffle=True)

rec = ae.predict(X_all[..., None], verbose=0).squeeze()
err = np.mean((X_all - rec) ** 2, axis=1)
# порог по 99-му перцентилю на train
thr = np.percentile(err[:train_len - w + 1], 99)

# совместим длины
pred = np.zeros(len(series), dtype=int)
pred[w-1:] = (err > thr).astype(int)
df["pred_lstm_ae"] = pred
eval_predictions(df["label"], df["pred_lstm_ae"], "LSTM-Autoencoder")
plot_predictions(df, "pred_lstm_ae", "LSTM-AE reconstruction error threshold (P99 на train)")
```

## 5. Предиктивная аналитика (forecasting)

Прогнозирование — вторая ключевая задача в мониторинге. Применений много:
оценка остаточного ресурса, capacity planning, формирование «нормального
коридора» для последующей детекции аномалий.

### Метрики качества прогноза

- **MAE** (Mean Absolute Error): $\frac{1}{n}\sum |y_t - \hat y_t|$
  — в тех же единицах, что и ряд; устойчива к выбросам.
- **RMSE** (Root Mean Squared Error): $\sqrt{\frac{1}{n}\sum (y_t - \hat y_t)^2}$
  — штрафует большие ошибки сильнее MAE.
- **MAPE** (Mean Absolute Percentage Error): $\frac{100}{n}\sum |y_t - \hat y_t| / |y_t|$
  — относительная, проблематична при $y_t \approx 0$.

### Валидация на временных рядах

Случайный train/test split **нельзя** — будет утечка из будущего в прошлое.
Используем простейший hold-out: первые 75% — обучение, последние 25% — тест.
Для серьёзных моделей применяют **walk-forward** валидацию (расширяющееся окно).

```python
def mae(y, yhat):   return float(np.mean(np.abs(y - yhat)))
def rmse(y, yhat):  return float(np.sqrt(np.mean((y - yhat) ** 2)))
def mape(y, yhat):
    mask = np.abs(y) > 1e-6
    return float(np.mean(np.abs((y[mask] - yhat[mask]) / y[mask])) * 100)

results_fc = []

def eval_forecast(y_true, y_pred, name):
    m_ = mae(y_true, y_pred); r_ = rmse(y_true, y_pred); p_ = mape(y_true, y_pred)
    print(f"{name:30s} | MAE={m_:.3f}  RMSE={r_:.3f}  MAPE={p_:.2f}%")
    results_fc.append({"model": name, "MAE": m_, "RMSE": r_, "MAPE": p_})
    return m_, r_, p_

# единый train/test split
split = int(len(df) * 0.75)
y_train = df["value"].values[:split]
y_test  = df["value"].values[split:]
ts_train = df["timestamp"].values[:split]
ts_test  = df["timestamp"].values[split:]
print(f"Train: {len(y_train)}, Test: {len(y_test)}")
```

### 5.1 Naive baselines

**Идея.** Простейшие правила, обязательные как точка отсчёта:
- **Last value**: $\hat y_{t+h} = y_t$
- **Mean baseline**: $\hat y_{t+h} = \bar y$
- **Seasonal naive**: $\hat y_{t+h} = y_{t+h-s}$ (значение из предыдущего сезона)

**Когда работает.** Если эти baselines дают F1/RMSE близкие к сложным моделям —
сложные модели здесь не нужны. Иначе можно понять верхнюю границу выигрыша.

**Ограничения.** Не моделируют изменения; для seasonal naive нужен корректный
период сезонности.

```python
y_pred_last = np.full(len(y_test), y_train[-1])
y_pred_mean = np.full(len(y_test), y_train.mean())
# seasonal naive — берём значения 1 период назад (для test берём из train или test-self)
y_pred_seasonal = np.array([
    df["value"].values[split + i - period] if split + i - period >= 0 else y_train.mean()
    for i in range(len(y_test))
])

eval_forecast(y_test, y_pred_last, "Naive: last value")
eval_forecast(y_test, y_pred_mean, "Naive: mean")
eval_forecast(y_test, y_pred_seasonal, "Naive: seasonal (1 day back)")
```

### 5.2 Holt-Winters (тройное экспоненциальное сглаживание)

**Идея.** Раскладывает ряд на три экспоненциально сглаженные компоненты:
уровень $\ell_t$, тренд $b_t$ и сезонность $s_t$ (аддитивная или
мультипликативная). Аддитивный вариант:

$$\ell_t = \alpha(y_t - s_{t-m}) + (1 - \alpha)(\ell_{t-1} + b_{t-1})$$
$$b_t   = \beta(\ell_t - \ell_{t-1}) + (1 - \beta) b_{t-1}$$
$$s_t   = \gamma(y_t - \ell_{t-1} - b_{t-1}) + (1 - \gamma) s_{t-m}$$
$$\hat y_{t+h} = \ell_t + h \cdot b_t + s_{t-m+1+((h-1) \mod m)}$$

**Преимущество.** Дёшево и хорошо работает на «гладких» рядах с явной
сезонностью.

**Ограничения.** Линейный тренд и одна сезонность; не учитывает экзогенные
переменные.

```python
hw_period = min(period, 96)  # ограничим для скорости
try:
    # trend=None: ряд CPU стационарен, аддитивный тренд на длинном горизонте
    # экстраполируется линейно и разваливает прогноз. Без тренда модель
    # описывает уровень + сезонность — это естественно для данных мониторинга.
    hw = ExponentialSmoothing(y_train, trend=None, seasonal="add",
                              seasonal_periods=hw_period).fit(optimized=True)
    y_pred_hw = hw.forecast(len(y_test))
    eval_forecast(y_test, y_pred_hw, "Holt-Winters (no trend, add seasonal)")
except Exception as e:
    print("Holt-Winters не построилась:", e)
    y_pred_hw = None
```

### 5.3 SARIMA

**Идея.** Параметрическая модель из главы 4.9, но теперь используем её для
многошагового прогноза.

**Преимущество.** Интерпретируемые коэффициенты; естественные доверительные
интервалы; хорошо работает на коротких рядах.

**Ограничения.** Подбор $(p,d,q,P,D,Q)$ нетривиален; долгое обучение на
длинных рядах; не масштабируется на десятки экзогенных признаков.

```python
sarima_fc_period = min(period, 24)
try:
    sm = SARIMAX(y_train, order=(1,1,1), seasonal_order=(1,1,1, sarima_fc_period),
                 enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)
    y_pred_sarima = sm.forecast(steps=len(y_test))
    eval_forecast(y_test, y_pred_sarima, "SARIMA(1,1,1)x(1,1,1)")
except Exception as e:
    print("SARIMA не построилась:", e)
    y_pred_sarima = None
```

### 5.4 Prophet (Facebook/Meta)

**Идея.** Аддитивная регрессионная модель: тренд $g(t)$, сезонные эффекты
$s(t)$ (через ряды Фурье), праздники $h(t)$ и шум $\varepsilon_t$. Обучается
байесовски (Stan).

$$y_t = g(t) + s(t) + h(t) + \varepsilon_t$$

**Преимущество.** Робастна к пропускам и выбросам; почти не требует
настройки; легко добавлять регрессоры и точки структурных сдвигов.

**Ограничения.** Чисто аддитивная декомпозиция; не учитывает автокорреляцию
остатков; на тонко-настроенных задачах уступает SARIMA и ML-моделям.

```python
pdf = df[["timestamp", "value"]].rename(columns={"timestamp": "ds", "value": "y"})
m = Prophet(daily_seasonality=True, weekly_seasonality=False, yearly_seasonality=False)
m.fit(pdf.iloc[:split])
fc = m.predict(pdf[["ds"]].iloc[split:])
y_pred_prophet = fc["yhat"].values
eval_forecast(y_test, y_pred_prophet, "Prophet")
```

### 5.5 LightGBM с временными признаками

**Идея.** Сводим прогнозирование к классической задаче регрессии.
Признаки для точки $t$ генерируются по прошлому:

- **Лаги**: $y_{t-1}, y_{t-2}, \ldots, y_{t-k}$
- **Скользящие статистики**: rolling mean, std, min/max за окно
- **Календарные**: час, день недели, минута часа

Обучаем градиентный бустинг (LightGBM) предсказывать $y_t$.

**Преимущество.** Лёгкий учёт многих признаков и регрессоров; хорошо
масштабируется; стабильно даёт сильные результаты на бенчмарках.

**Ограничения.** Прогноз на много шагов — рекурсивный или прямой —
накапливает ошибку; модель не «знает» про физику ряда.

```python
def build_features(values, ts, lags=(1, 2, 3, 6, 12, 24)):
    s = pd.Series(values, index=pd.DatetimeIndex(ts))
    feat = pd.DataFrame(index=s.index)
    for L in lags:
        feat[f"lag_{L}"] = s.shift(L)
    feat["roll_mean_6"]  = s.shift(1).rolling(6).mean()
    feat["roll_std_6"]   = s.shift(1).rolling(6).std()
    feat["roll_mean_24"] = s.shift(1).rolling(24).mean()
    feat["hour"]   = s.index.hour
    feat["dow"]    = s.index.dayofweek
    feat["minute"] = s.index.minute
    feat["y"] = s.values
    return feat.dropna()

full_ts = df["timestamp"].values
full_y  = df["value"].values

F = build_features(full_y, full_ts)
F_train = F.loc[F.index < df["timestamp"].iloc[split]]
F_test  = F.loc[F.index >= df["timestamp"].iloc[split]]
X_tr, y_tr = F_train.drop(columns="y"), F_train["y"]
X_te, y_te = F_test.drop(columns="y"),  F_test["y"]

model = lgb.LGBMRegressor(n_estimators=300, learning_rate=0.05, num_leaves=31,
                          min_child_samples=20, random_state=42, verbose=-1)
model.fit(X_tr, y_tr)
y_pred_lgbm = model.predict(X_te)
eval_forecast(y_te.values, y_pred_lgbm, "LightGBM (lag features)")
```

### 5.6 LSTM Forecaster

**Идея.** Рекуррентная нейросеть на вход получает окно длины $w$ значений,
на выходе — следующее значение:

$$\hat y_t = f_\theta(y_{t-w}, \ldots, y_{t-1})$$

Обучается оптимизацией MSE; внутри — LSTM-ячейки с gating-механизмом
(input/forget/output gates), благодаря которому модель удерживает
долгосрочные зависимости.

**Преимущество.** Захватывает нелинейные зависимости и сложные паттерны.

**Ограничения.** Чёрный ящик; требует много данных и подбора гиперпараметров;
дороже в эксплуатации (GPU, инференс-латентность); легко переобучается.

```python
from tensorflow.keras import layers, models
tf.random.set_seed(42)

w = 24
series = df["value"].values.astype("float32")
mu_, sd_ = series[:split].mean(), series[:split].std() + 1e-9
norm = (series - mu_) / sd_

def make_xy(arr, w):
    X = np.stack([arr[i:i+w] for i in range(len(arr) - w)])
    y = arr[w:]
    return X, y

X_full, y_full = make_xy(norm, w)
n_train = split - w
X_tr_n, y_tr_n = X_full[:n_train], y_full[:n_train]
X_te_n, y_te_n = X_full[n_train:], y_full[n_train:]

inp = layers.Input(shape=(w, 1))
x = layers.LSTM(32)(inp)
out = layers.Dense(1)(x)
lstm = models.Model(inp, out)
lstm.compile(optimizer="adam", loss="mse")
lstm.fit(X_tr_n[..., None], y_tr_n, epochs=10, batch_size=64, verbose=0)

yhat_norm = lstm.predict(X_te_n[..., None], verbose=0).squeeze()
y_pred_lstm = yhat_norm * sd_ + mu_
y_true_lstm = y_te_n * sd_ + mu_
eval_forecast(y_true_lstm, y_pred_lstm, "LSTM forecaster")
```

## 6. Forecast-based anomaly detection

**Идея.** Любая forecasting-модель даёт нам ожидаемое значение
$\hat y_t$ и оценку шумовой дисперсии $\sigma$. Тогда правило детекции
аномалий — выход за «нормальный коридор»:

$$\text{аномалия, если } |y_t - \hat y_t| > k \cdot \sigma$$

или эквивалентно — точка выходит за prediction interval. Преимущество подхода —
модель ловит **контекстные** аномалии: значение может быть само по себе
обычным, но необычным для данного момента дня/недели.

Применим это к прогнозам Holt-Winters (для всего теста — sliding 1-step не делаем
ради скорости; для иллюстрации хватит сравнения тестовых ошибок с
дисперсией шума на train).

```python
# Используем in-sample one-step-ahead предсказания: пере-фитим HW
# на ВЕСЬ ряд и берём fittedvalues. В продакшене эквивалент — rolling
# one-step forecast с регулярной дообучкой модели.
try:
    hw_full = ExponentialSmoothing(df["value"].values, trend=None, seasonal="add",
                                   seasonal_periods=hw_period).fit(optimized=True)
    yhat = hw_full.fittedvalues
    resid = df["value"].values - yhat
    sigma = np.std(resid[:int(len(resid)*0.5)])  # оценка σ по «нормальной» половине
    k_thr = 3.0
    df["pred_fc_based"] = (np.abs(resid) > k_thr * sigma).astype(int)
    eval_predictions(df["label"], df["pred_fc_based"], "Forecast-based (HW one-step, 3σ)")
    plot_predictions(df, "pred_fc_based", "Forecast-based AD via Holt-Winters (one-step in-sample)")
except Exception as e:
    print("Forecast-based AD не построилась:", e)
```

## 7. Многомерный временной ряд (мультивариативный анализ)

В реальном мониторинге обычно отслеживают сразу много метрик: CPU, память,
сеть, IOPS и т.д. — на каждом из множества хостов. Аномалия может проявляться
**не в каждой отдельной метрике**, а в их совместном поведении (например,
рост CPU при одновременном падении сетевого трафика).

В этом блоке мы возьмём CPU-метрики пяти разных EC2-инстансов из NAB и
рассмотрим их как многомерный временной ряд «парка серверов». Так как в
NAB разные файлы соответствуют разным временным диапазонам, для целей
наглядности мы перенумеруем точки на единую целочисленную ось (по индексу
наблюдения), сохранив исходные метки времени отдельно.

Применим два многомерных подхода:

- **Расстояние Махаланобиса** — устойчивая «дистанция от центра» с учётом
  ковариационной структуры.
- **Isolation Forest** на многомерных признаках.

Визуализации будем строить интерактивно с помощью Plotly.

```python
# Загружаем 5 EC2 CPU метрик в общий DataFrame
multi_dfs = []
for path_rel in MULTIVARIATE_FILES:
    local = path_rel.replace("/", "_")
    dfi = pd.read_csv(local, parse_dates=["timestamp"]).sort_values("timestamp").reset_index(drop=True)
    name = path_rel.split("/")[-1].replace(".csv", "")
    dfi = dfi.rename(columns={"value": name})
    multi_dfs.append(dfi.set_index("timestamp")[[name]])

# Каждый ряд приводим к общей длине = минимум, и кладём на синтетический индекс
min_len = min(len(d) for d in multi_dfs)
multi = pd.DataFrame({d.columns[0]: d.iloc[:min_len, 0].values for d in multi_dfs})
multi.index.name = "step"
print("Размер многомерного датасета:", multi.shape)
multi.head()
```

### 7.1 Расстояние Махаланобиса

**Идея.** Для многомерных нормальных данных хорошим скором аномальности
является расстояние Махаланобиса от центра распределения:

$$D_M(x) = \sqrt{(x - \mu)^\top \Sigma^{-1} (x - \mu)}$$

где $\mu, \Sigma$ — оценки среднего вектора и ковариационной матрицы по
«нормальным» данным. Под гипотезой нормальности $D_M^2$ распределено как
$\chi^2_d$ (где $d$ — размерность); пороги выбираются по квантилям этого
распределения.

**Преимущество.** Учитывает корреляции между признаками: точка, лежащая
далеко **поперёк главных направлений ковариационной матрицы**, считается
аномальной даже если по отдельным осям она в пределах нормы.

**Ограничения.** Чувствителен к выбросам в оценке $\Sigma$ (используем
ковариацию на «чистой» половине данных или робастные оценки, например
MinCovDet); требует нормальности; сложно интерпретировать, какая именно
метрика «виновата».

```python
from scipy.stats import chi2

X = multi.values.astype(float)
n_train_m = int(len(X) * 0.5)

mu_vec = X[:n_train_m].mean(axis=0)
cov = np.cov(X[:n_train_m].T)
inv_cov = np.linalg.pinv(cov)

diff = X - mu_vec
D2 = np.einsum("ij,jk,ik->i", diff, inv_cov, diff)
d = X.shape[1]
threshold = chi2.ppf(0.999, df=d)

multi["mahalanobis_d2"] = D2
multi["mahal_anomaly"] = (D2 > threshold).astype(int)
print(f"Порог χ² (df={d}, 99.9%): {threshold:.2f}")
print(f"Обнаружено аномалий: {int(multi['mahal_anomaly'].sum())} из {len(multi)}")

fig = go.Figure()
fig.add_trace(go.Scatter(x=multi.index, y=D2, mode="lines", name="D²(Mahalanobis)",
                         line=dict(color="purple", width=1)))
fig.add_hline(y=threshold, line=dict(color="red", dash="dash"),
              annotation_text=f"χ² 99.9% = {threshold:.1f}", annotation_position="top right")
anom_idx = multi[multi["mahal_anomaly"] == 1].index
fig.add_trace(go.Scatter(x=anom_idx, y=D2[anom_idx], mode="markers",
                         name="Аномалии", marker=dict(color="red", size=6)))
fig.update_layout(title="Расстояние Махаланобиса по многомерному ряду",
                  template="plotly_white", height=420,
                  xaxis_title="step", yaxis_title="D²",
                  hovermode="x unified")
fig.show()
```

### 7.2 Многомерный Isolation Forest

Применим Isolation Forest напрямую к многомерным признакам.
Дополнительно подадим скользящие средние по каждой метрике, чтобы модель
имела «контекст» окрестности.

```python
feat_multi = multi[[c for c in multi.columns if c not in ("mahalanobis_d2", "mahal_anomaly")]].copy()
for col in list(feat_multi.columns):
    feat_multi[f"{col}__rmean24"] = feat_multi[col].rolling(24, min_periods=1).mean()

iso_m = IsolationForest(contamination=0.02, random_state=42, n_estimators=200)
multi["iforest_anomaly"] = (iso_m.fit_predict(feat_multi.values) == -1).astype(int)
print(f"Multivariate IsolationForest аномалий: {int(multi['iforest_anomaly'].sum())}")
```

## 8. Итоги и обсуждение

**Что мы попробовали:**

| Блок | Методы | Главная идея |
|------|--------|--------------|
| Статистические | Z-score, MAD, IQR, rolling 3σ | пороги по уровню или локальной норме |
| Декомпозиционные | STL + порог по остатку | убираем тренд и сезонность, ловим выбросы остатка |
| ML без учителя | Isolation Forest, LOF, One-Class SVM | геометрия и плотность в пространстве признаков |
| Forecast-based | SARIMA, Prophet, Holt-Winters + интервал | контекстные аномалии — отклонения от прогноза |
| Deep Learning | LSTM-Autoencoder | реконструкционная ошибка нормальных паттернов |
| Multivariate | Mahalanobis, multivariate IsolationForest | совместное поведение нескольких метрик |

**Практические выводы для мониторинга:**

1. **Начинайте с простого.** Часто Z-score, MAD, rolling 3σ и STL покрывают
   80% случаев и дают предсказуемый порог срабатывания. Они объяснимы и
   почти не требуют поддержки.

2. **Контекстные аномалии** (когда значение нормально само по себе, но
   нетипично для текущей фазы цикла) удобнее ловить через forecast-based
   подходы (Prophet/SARIMA/HW + prediction interval).

3. **ML без учителя** (Isolation Forest, LOF, OCSVM) хорошо работает в
   многомерных задачах и при отсутствии разметки, но требует подбора
   `contamination` и качественных признаков (включая скользящие
   статистики и остатки декомпозиции).

4. **Deep learning** (LSTM-AE и т.п.) даёт преимущество на сложных
   нелинейных паттернах с длинной памятью, но дорог в эксплуатации и
   нуждается в значительном объёме «нормальных» данных.

5. **Trade-off precision vs recall.** В SOC и в АСУ ТП один FP — это
   ложный вызов смены; FN — пропуск инцидента. Стоимость FP/FN формирует
   выбор порога; для этого NAB и вводит специальную метрику NAB Score.

6. **Production-аспекты.** Учитывайте concept drift (модели нужно
   перетренировывать), latency онлайн-скоринга и observability самих
   моделей (например, через Evidently/WhyLabs — в курсе они уже
   рассматривались).

**Куда двигаться дальше:**

- сравнить полученные результаты с эталонной NAB Score;
- добавить ансамблирование детекторов (мажоритарное голосование);
- попробовать многомерные методы на реально многомерных данных
  (Server Machine Dataset, SMAP/MSL);
- закрыть пайплайн с retraining и алертингом в production-стиле.

**Важная особенность разметки NAB.** В `combined_windows.json` аномалии заданы широкими **окнами** (~10% точек ряда), которые включают предшествующий аномалии контекст — это нужно для метрики **NAB Score**, поощряющей раннее обнаружение. Точечные детекторы (Z-score, MAD, SARIMA-interval) обычно показывают **высокую precision** (они ловят сам спайк), но **низкий recall и F1** по точечной метрике, потому что не помечают весь контекст. Это нормально и хорошо согласуется с продакшеновой логикой: важно срабатывать на инциденте, а не «закрашивать» все часы вокруг. Для академической оценки лучше использовать оригинальный NAB Score или ослабленный point-adjustment.
